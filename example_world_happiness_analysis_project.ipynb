# Data Analysis Project: Understanding World Happiness
This project explores the World Happiness Report dataset to uncover what factors contribute most to the happiness scores of countries around the world.
## Setup and Data Loading
First, we need to set up our environment and load the dataset into a Pandas DataFrame.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Load the dataset
df = pd.read_csv('./data/whr_2019.csv')

# Display the first few rows of the dataframe
df.head()
```

## Data Exploration
Let's explore the dataset by checking its structure, summary statistics, and identifying any initial trends or anomalies.

```python
# Check the structure and summary statistics of the dataset
df.info()
df.describe()
```

## Data Cleaning and Preparation
Before diving into analysis, let's prepare our data by handling missing values and ensuring correct data types.

```python
# Set 'Country or region' as the index of the DataFrame
df.set_index('Country or region', inplace=True)
```

## Data Analysis and Visualization
Now, let's dive into the data analysis, focusing on understanding how different factors contribute to happiness scores. We'll use both statistical analysis and visualization techniques.

```python
# Analyzing relationships between happiness score and factors
# Correlation analysis
correlation_matrix = df.corr()
sns.heatmap(correlation_matrix, annot=True)
plt.title('Correlation Analysis of Happiness Score and Contributing Factors')
plt.show()

# Visualization of relationships
sns.pairplot(df, vars=['Score', 'GDP per capita', 'Social support', 'Healthy life expectancy', 'Freedom to make life choices', 'Generosity', 'Perceptions of corruption'])
plt.show()
```
